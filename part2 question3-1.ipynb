- Activity and resource are aggregated using frequency
- Timestamp are aggregated using average

```python
import pandas as pd
import numpy as np
```

```python
log_bpic = pd.read_csv("log_bpic.csv", index_col=0)
log_bpic
```

```python
log2_bpic = log_bpic
```

```python
log2_bpic['ts'] = pd.to_datetime(log_bpic['ts'], format='%Y/%m/%d %H:%M:%S.%f')
log2_bpic
```

```python
# caseid 별로 average ts 데이터프레임 만들기
df = log2_bpic.groupby(['caseid'])['ts'].mean()
df
```
