6.	Train a decision tree and a randomforest from the scikit-learn package for outcome label prediction using the event log created at step 2 and and step 5 for prefix length L=2 and L=5.  Split 70-30 for training-testing.

Hyperparameters
DT: max_depth=5
RF: n_estimators=100, oob_score=True


```python
#step 2 decision tree
import pandas as pd
import numpy as np

from sklearn import datasets
from sklearn import tree
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
```

```python
log_bpic = pd.read_csv("log_bpic.csv", index_col=0)
log_bpic
```

```python
bpic_data = log_bpic[["caseid", "activity", "ts", "resource"]]
bpic_data
```

```python
bpic_data_caseid = bpic_data[["caseid"]]
bpic_data_caseid
```

```python
bpic_data_activity = bpic_data[["activity"]]
bpic_data_activity
```

```python
bpic_data_ts = bpic_data[["ts"]]
bpic_data_ts
```

```python
bpic_data_resource = bpic_data[["resource"]]
bpic_data_resource
```

```python
#how to use string data in scikit-learn DT
#https://datascience.stackexchange.com/questions/5226/strings-as-features-in-decision-tree-random-forest

#one-hot encoding
#https://injo.tistory.com/11

from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import LabelEncoder

encoder = LabelEncoder()
encoder.fit(bpic_data_caseid)
labels_caseid = encoder.transform(bpic_data_caseid)
labels_caseid = labels_caseid.reshape(-1,1)

print('인코딩 변화값: ', labels_caseid)
```

```python
encoder = LabelEncoder()
encoder.fit(bpic_data_activity)
labels_activity = encoder.transform(bpic_data_activity)
labels_activity = labels_activity.reshape(-1,1)

print('인코딩 변화값: ', labels_activity)
```

```python
encoder = LabelEncoder()
encoder.fit(bpic_data_ts)
labels_ts = encoder.transform(bpic_data_ts)
labels_ts = labels_ts.reshape(-1,1)

print('인코딩 변화값: ', labels_ts)
```

```python
encoder = LabelEncoder()
encoder.fit(bpic_data_resource)
labels_resource = encoder.transform(bpic_data_resource)
labels_resource = labels_resource.reshape(-1,1)

print('인코딩 변화값: ', labels_resource)
```

```python
bpic_target = log_bpic[["outcome"]]
bpic_target
```

```python
# 원핫인코딩 1d 데이터들 합치기
bpic_data_encode= np.concatenate((labels_caseid, labels_activity, labels_ts, labels_resource), axis=1)
bpic_data_encode
```

```python
X_train, X_test, y_train, y_test = train_test_split(bpic_data_encode, bpic_target, test_size=0.3, random_state=10)
```

```python
step2_model_dt = DecisionTreeClassifier(max_depth = 5)
```

```python
step2_model_dt.fit(X_train, y_train)

print("train score: {:.3f}".format(step2_model_dt.score(X_train, y_train)))

print("test score: {:.3f}".format(step2_model_dt.score(X_test, y_test)))
```

```python
#step 2 randomforest

import pandas as pd
import numpy as np

from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
```

```python
model = RandomForestClassifier(n_estimators=5, random_state=0, oob_score=True)
model.fit(X_train, y_train)
```

```python
# 모델 학습
model = RandomForestClassifier(n_estimators=10, random_state=0,
                              max_features=4, oob_score=True)
model.fit(X_train, y_train)

# 평가
print("훈련 세트 정확도: {:.3f}".format(model.score(X_train, y_train)) )
print("테스트 세트 정확도: {:.3f}".format(model.score(X_test, y_test)) )
print("OOB 샘플의 정확도: {:.3f}".format(model.oob_score_) )
```
